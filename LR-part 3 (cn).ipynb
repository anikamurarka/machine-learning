```python
import numpy as np
data_test = np.loadtxt("0000000000002329_test_diabetes_x_test.csv",delimiter=",")

x_test = data_test[:,0]
y_test = data_test [:,1]
data_test.shape
```

```python
data_train = np.loadtxt("0000000000002329_training_diabetes_x_y_train.csv",delimiter=",")
x_train = data_train[:,0]
y_train = data_train[:,1]
y_train
```

```python
def fit (x_train,y_train):
    xy_mean = (x_train*y_train).mean()
    x_mean = x_train.mean()
    y_mean = y_train.mean()
    x2_mean = (x_train**2).mean()
    m = (xy_mean - (x_mean * y_mean))/(x2_mean -x_mean )
    c = y_mean - m*x_mean
    return m,c
```

```python
m,c = fit(x_train,y_train)
```

```python
def predict(x,m,c):
    return m*x + c
```

```python
y_pred = predict(x_test,m,c)
```

```python
def score(y_pred,y_test):
    u = ((y_test-y_pred)**2).sum()
    v = ((y_test-(y_test).mean())**2).sum()
    return 1 - (u/v)
```

```python
score(y_pred,y_test)
```
