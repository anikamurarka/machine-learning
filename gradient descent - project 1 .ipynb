```python
def cost(m, data):
    M = data.shape[0]
#     X = data[]
    num_feat = data.shape[1]-1
    cst=0
    for i in range(M):
        cst+=(1/M)*((data[i][num_feat]-sum(m*data[:,:num_feat][i]))**2)
    print("cost",cst)
```

```python
def step(data,L_R,m):
    M = data.shape[0]
#     print("M",M)
    num_feat = data.shape[1]-1
    np.set_printoptions(formatter={'float_kind':'{:f}'.format})
#     print("num_feat",num_feat)
    slope_m = np.zeros((data.shape[1]-1,),int)
    for i in range(M):
        x=data[:,:num_feat][i]
        y=data[i][num_feat]
        for j in range(len(slope_m)):
            slope_m[j]+=(-2)*(y-((m*x).sum()))*x[j]
#     for k in range(num_feat):
#         for i in range(M):
#             y = data[i][num_feat]
# #             print("y",y)
#             inner_calc = 0
#             for j in range(num_feat):
#                 x = data[i][j]
# #                 print("x",x)
#                 inner_calc += m[j]*x
# #             print(y,inner_calc,data[i][k],M)
#             slope_m[k] += -((y - inner_calc)*data[i][k]*2)/M
# #             print("slope",slope_m)
    new_m = m - L_R*slope_m
    cost(m,data)
    return new_m
```

```python
def gd2(data,L_R,no_of_iterations):
    m = np.zeros((data.shape[1]-1,),int)
#     print(data.shape)
#     print(m.shape)
    iterations = no_of_iterations
    while no_of_iterations>0:
        m = step(data,L_R,m)
        no_of_iterations -= 1
    return m
```
